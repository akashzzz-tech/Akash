```
feature_col=['TV','Radio','Newspaper']
x=sales_data[feature_col]
y=sales_data.Sales
```

```
x_train, x_test, y_train, y_test = train_test_split(x,y,
  test_size=0.2,random_state= 42)
```

```
display(x_train.shape, x_test.shape, y_train.shape ,y_test.shape)
```

```
print(x_train)
print(y_train)
```

```
model=LinearRegression()
model.fit(x_train,y_train)    #training phase
y_pred=model.predict(x_test)  #testing phase by giving input
```

```
print(y_pred)
print(x_test)
```

```
#feature = [[TV,Radio,Newspaper]]
features = np.array([[230,150,150]])
print(model.predict(features))
```

```
from sklearn.metrics import mean_absolute_error , mean_squared_error,mean_squared_log_error, r2_score
```

```
#r-sqr check
print("R-sqr value:",r2_score(y_pred,y_test))
```

```
print("means absolute error rate:",mean_absolute_error(y_pred,y_test))
```

```
print("means squared error rate:",mean_squared_error(y_test,y_pred))
```
